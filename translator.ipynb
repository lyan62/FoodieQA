```python
import deepl
import json
import torch
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
def read_data(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
        questions = [i["question"] for i in data]
        choices = [j for i in data for j in i["choices"]]
        return data, questions, choices

# file_path = "./dataset/mivqa_filtered.json"
file_path = "./dataset/sivqa_filtered.json"
raw_data, questions, choices = read_data(file_path)

len(raw_data), len(questions), questions[:5], choices
```

```python
def translate(api_name, source_text):
    if api_name == "deepl":
        api_key = os.getenv("DEEPL_API_KEY")
        # print(api_key)
        translator = deepl.Translator(api_key)
        result = translator.translate_text(source_text, source_lang="ZH", target_lang="EN-US")
    return result

translated_q = translate("deepl", questions)
testtranslated_c = translate("deepl", choices)
```

```python
raw_translated_question = [t.text for t in translated_q]
raw_translated_choices = [t.text for t in testtranslated_c]

# with open("translated_questions_en.json", "w") as file:
#     json.dump(raw_translated_question, file, indent=4)
```
